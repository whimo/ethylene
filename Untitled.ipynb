```python
import numpy as np
import pandas
```

```python
data = pandas.read_csv('data_for_MIPT.csv')
data.head()
```

```python
train = data.drop(['DateTime', 'target'], axis=1)
target = data['target']
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, target, test_size=0.3)
```

```python
from catboost import CatBoostRegressor
```

```python
cb = CatBoostRegressor(task_type='GPU')
cb.fit(X_train, y_train)
```

```python
print(np.mean(cross_val_score(CatBoostRegressor(task_type='GPU', n_estimators=5000, depth=3), train, target, cv=5, scoring='r2')))
```

```python
from sklearn.linear_model import Lasso
```

```python
print(np.mean(cross_val_score(Lasso(), train, target, cv=5, scoring='r2')))
```
